```python
# library = ast and function = ast.literal_eval helps to convert string format to integer.
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
 def convert(obj): #calling the helper function
     L=[]          #making a list
     for i in ast.literal_eval(obj):
         L.append(i['name'])
     return L   
```

```python
def convert3(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
movies['cast'].apply(convert3)
```

```python
movies['cast']= movies['cast'].apply(convert3)
```

```python
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'].apply(fetch_director)
```

```python
movies['crew']= movies['crew'].apply(fetch_director)
```

```python
import nltk #nltk is a natural language processing library
```

```python
from nltk.stem.porter import PorterStemmer #PorterStemmer is a class
ps = PorterStemmer()  #making object of PorterStemmer as ps
```

```python
def stem(text):  #creating a helper function
    y = []       #creating a list  

    for i in text.split():  #converting string into list
        y.append(ps.stem(i))

    return " ".join(y) #converting this list back into string
```

```python
new_df['tags'].apply(stem)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse = True, key=lambda x:x[1])[1:6]

    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```
